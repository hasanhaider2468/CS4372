```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock_symbol = 'KO'  # using coke for my example
start_date = '1993-01-01'
end_date = '2023-12-31'

stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

stock_data.head()
```

```python
# Parameters for initial investment and SMA windows
initial_investment = 10000  
short_term_window = 50  
long_term_window = 200  


short_term_sma = stock_data['Close'].rolling(window=short_term_window).mean()
long_term_sma = stock_data['Close'].rolling(window=long_term_window).mean()

# Buy and Hold Strategy
buy_and_hold_equity = []

# tracking the growth of the investment from the start
for month in range(len(stock_data)):
    equity = initial_investment * (stock_data['Close'].iloc[month] / stock_data['Close'].iloc[0]) 
    #end of month closing data
    buy_and_hold_equity.append(equity)

active_trading_investment = 0
in_stock = False
equity_at_month_list = []

for month in range(long_term_window, len(stock_data)):
    short_term_sma_value = short_term_sma.iloc[month].iloc[0]
    long_term_sma_value = long_term_sma.iloc[month].iloc[0]  
    
    # Buy signal: Short-term SMA crosses above long-term SMA
    if short_term_sma_value > long_term_sma_value and not in_stock:
        active_trading_investment += initial_investment  # Buy stock
        in_stock = True
    
    # Sell signal: Short-term SMA crosses below long-term SMA
    elif short_term_sma_value < long_term_sma_value and in_stock:
        active_trading_investment += initial_investment  # Sell stock
        in_stock = False

    # Calculating equity at current month
    equity_at_month = active_trading_investment * (stock_data['Close'].iloc[month] / stock_data['Close'].iloc[0])
    
    #lastly we append equity
    equity_at_month_list.append(equity_at_month)

# this is just preperation for plotting
buy_and_hold_equity = np.array(buy_and_hold_equity)
equity_at_month_list = np.array(equity_at_month_list)
```

```python
# Plotting the results
plt.figure(figsize=(10,6))
plt.plot(range(len(stock_data)), buy_and_hold_equity, label='Buy and Hold Strategy', color='green')
plt.plot(range(long_term_window, len(stock_data)), equity_at_month_list, label='Active Trading Strategy', color='blue')
plt.xlabel('Months')
plt.ylabel('Portfolio Value')
plt.title(f'{stock_symbol} - Buy and Hold vs Active Trading Strategy (SMA Crossover)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# determining the total equity by taking the last values
total_equity_bah = buy_and_hold_equity[-1]
total_equity_at = equity_at_month_list[-1]

# getting final equity
final_equity_bah = total_equity_bah[-1] if isinstance(total_equity_bah, (np.ndarray, list)) else total_equity_bah
final_equity_at = total_equity_at[-1] if isinstance(total_equity_at, (np.ndarray, list)) else total_equity_at

# taking annual return percentages by dividing by the year (time) and initial investment
annual_return_bah = (final_equity_bah / initial_investment) ** (1 / (len(stock_data) / 12)) - 1
annual_return_at = (final_equity_at / initial_investment) ** (1 / (len(stock_data) / 12)) - 1

# Print results
print(f"Buy and Hold - Total Equity: ${final_equity_bah:.2f}, Annual Return: {annual_return_bah*100:.2f}%")
print(f"Active Trading - Total Equity: ${final_equity_at:.2f}, Annual Return: {annual_return_at*100:.2f}%")
```
