```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255, 
                                   rotation_range=40, 
                                   width_shift_range=0.2, 
                                   height_shift_range=0.2, 
                                   shear_range=0.2, 
                                   zoom_range=0.2,
                                   horizontal_flip=True, 
                                   validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    'RealWaste',  #this our data
    target_size=(224, 224), #using this as our pixel size for the target wastw
    batch_size=32, #batch size can increase but can be computationally heavy
    class_mode='categorical', 
    subset='training'
)

validation_generator = train_datagen.flow_from_directory( #same idea as before except validation subset
    'RealWaste',  
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras import models, layers
from tensorflow.keras.applications import MobileNetV2

# Function to create model with variable layer count and dropout rate
def create_model(layer_count, dropout_rate):
    base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
    base_model.trainable = False

    model = models.Sequential()
    model.add(base_model)
    model.add(layers.GlobalAveragePooling2D())

    # Add variable number of Dense layers with ReLU / relu is the simplest and commonly used
    for _ in range(layer_count):
        model.add(layers.Dense(256, activation='relu'))
        model.add(layers.Dropout(dropout_rate))

    model.add(layers.Dense(9, activation='softmax'))  # last layer

    return model

# DataFrame to log experiments / Here I can run multiple iterations with different layers/dropouts
log_data = {
    "Iteration": [1, 2, 3, 4, 5, 6],
    "Parameters": ["Layer Count: 3, Dropout: 0.5", "Layer Count: 4, Dropout: 0.25", "Layer Count: 5, Dropout: 0.3",
                   "Layer Count: 3, Dropout: 0.4", "Layer Count: 4, Dropout: 0.35", "Layer Count: 5, Dropout: 0.2"],
    "Training Accuracy": [None] * 6,
    "Validation Accuracy": [None] * 6
}

# Iterating
for i in range(6):
    if i == 0:
        model = create_model(layer_count=3, dropout_rate=0.5)
    elif i == 1:
        model = create_model(layer_count=4, dropout_rate=0.25)
    elif i == 2:
        model = create_model(layer_count=5, dropout_rate=0.3)
    elif i == 3:
        model = create_model(layer_count=3, dropout_rate=0.4)
    elif i == 4:
        model = create_model(layer_count=4, dropout_rate=0.35)
    else:
        model = create_model(layer_count=5, dropout_rate=0.2)

    # Compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    history = model.fit(train_generator, validation_data=validation_generator, epochs=10, batch_size=32)

    # Log training and validation accuracy for this iteration to save for later
    log_data["Training Accuracy"][i] = history.history['accuracy'][-1]
    log_data["Validation Accuracy"][i] = history.history['val_accuracy'][-1]

    # Plotting all the values
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy') #using acc
    plt.plot(history.history['val_accuracy'], label='Val Accuracy') #using validation acc
    plt.title(f'Iteration {i+1}: Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='upper left')

    # Plotting the training/validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'Iteration {i+1}: Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(loc='upper left')

    plt.show()

#here we create the actual df
df = pd.DataFrame(log_data)
print(df)
```
